```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Create a synthetic dataset with missing values
data = {
    'Horsepower': [100, 150, 200, 250, 300, 350, np.nan],  # Engine power
    'Weight': [2000, 2500, 3000, 3500, 4000, np.nan, 5000],  # Vehicle weight
    'Cylinders': [4, 4, 6, 6, 8, 8, 8],  # Number of engine cylinders
    'MPG': [30, 25, 20, 18, 15, 13, 10]  # Fuel efficiency (target)
}

# Convert data into a pandas DataFrame
df = pd.DataFrame(data)
```

```python
# Step 2: Initial data overview
print("Initial Dataset:")
print(df)
print("\nSummary Statistics:")
print(df.describe())

# Visualizing missing data
print("\nChecking for Missing Values:")
print(df.isnull().sum())
```

```python
# Step 3: Data visualization
# Pairplot for relationships between variables
sns.pairplot(df, diag_kind='kde')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()

# Histograms for each feature
df.hist(bins=10, figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", y=0.95)
plt.show()

# Boxplot to detect outliers
sns.boxplot(data=df, palette="pastel")
plt.title("Boxplots of Features")
plt.show()
```

```python
# Step 4: Handling missing data
# Using mean imputation for missing values
imputer = SimpleImputer(strategy='mean')
df[['Horsepower', 'Weight']] = imputer.fit_transform(df[['Horsepower', 'Weight']])

print("\nDataset After Imputation:")
print(df)
```

#### Use SimpleImputer for numerical columns with some missing values when you want to retain most of the data.
#### Avoid imputation for columns with excessive missing data or if the imputed value would distort the data's meaning.
#### You can apply alternative strategies (like row/column deletion or advanced imputation techniques) based on the situation.

```python
# Step 5: Normalizing numerical data
scaler = StandardScaler()
df[['Horsepower', 'Weight']] = scaler.fit_transform(df[['Horsepower', 'Weight']])

print("\nDataset After Normalization:")
print(df)
```

```python
# Step 6: Visualizing normalized data
# Scatterplot of normalized features
plt.scatter(df['Horsepower'], df['Weight'], c='blue', label='Normalized Data')
plt.xlabel("Normalized Horsepower")
plt.ylabel("Normalized Weight")
plt.title("Scatterplot of Normalized Features")
plt.legend()
plt.show()
```

```python
# Step 7: Correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

print("\nCorrelation Matrix:")
print(correlation_matrix)
```
