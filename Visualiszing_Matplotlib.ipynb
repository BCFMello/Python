```python
# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (you can adjust this to your own dataset)
file_path = '/content/Salaries.csv'
salaries_df = pd.read_csv(file_path)

# Renaming columns to make them more user-friendly (if not already renamed)
salaries_df.rename(columns={'EmployeeName': 'Employee Name', 'JobTitle': 'Job Title', 'TotalPay': 'Total Pay'}, inplace=True)

# Ensure numeric data is correct for plotting
salaries_df['BasePay'] = pd.to_numeric(salaries_df['BasePay'], errors='coerce')
salaries_df['Total Pay'] = pd.to_numeric(salaries_df['Total Pay'], errors='coerce')

# 44. Organize Plots using Subplot
# Creating multiple subplots for different visualizations
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Subplot 1: Histogram
axs[0, 0].hist(salaries_df['Total Pay'], bins=30, color='blue', edgecolor='black')
axs[0, 0].set_title('Histogram of Total Pay')

# Subplot 2: Pie chart
job_title_counts = salaries_df['Job Title'].value_counts().head(5)
axs[0, 1].pie(job_title_counts, labels=job_title_counts.index, autopct='%1.1f%%', startangle=90)
axs[0, 1].set_title('Pie Chart of Top 5 Job Titles')

# Subplot 3: Bar chart
axs[1, 0].bar(job_title_counts.index, job_title_counts.values, color='orange')
axs[1, 0].set_title('Bar Chart of Top 5 Job Titles')
axs[1, 0].tick_params(axis='x', rotation=45)

# Subplot 4: Scatter plot (Corrected the string error)
axs[1, 1].scatter(salaries_df['BasePay'], salaries_df['Total Pay'], color='green', alpha=0.5)
axs[1, 1].set_title('Scatter Plot of BasePay vs Total Pay')
axs[1, 1].set_xlabel('Base Pay')
axs[1, 1].set_ylabel('Total Pay')

plt.tight_layout()
plt.show()
```
